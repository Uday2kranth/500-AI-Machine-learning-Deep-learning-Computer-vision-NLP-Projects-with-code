```python
# Install dependencies (optional if already installed)

# Step 1: Train the model
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import pickle
```

```python

iris = load_iris()
x = iris.data
y = iris.target

model = DecisionTreeClassifier()
model.fit(x, y)

# Optionally save the model (not needed unless exporting)
with open("iris_model.pkl", "wb") as f:
  pickle.dump(model, f)

# Class labels
target_names = iris.target_names
```

```python
import gradio as gr
```

```python
def predict_iris(sepal_length, sepal_width, petal_length, petal_width):
  input_data=[[sepal_length, sepal_width, petal_length, petal_width]]
  prediction = model.predict(input_data)[0]
  return f"ðŸŒ¸ Predicted Species: **{target_names[prediction].capitalize()}**"

# Create Gradio UI
iface = gr.Interface(
    fn=predict_iris,
    inputs=[
        gr.Number(label="Sepal Length (cm)", value=5.1),
        gr.Number(label="Sepal Width (cm)", value=3.5),
        gr.Number(label="Petal Length (cm)", value=1.4),
        gr.Number(label="Petal Width (cm)", value=0.2),
    ],
    outputs="markdown",
    title="ðŸª» Iris Flower Classifier",
    description="Enter flower measurements to predict its species using a Decision Tree model.",
    theme="soft"
    )

# Launch the app
iface.launch()
```
